```python
import matplotlib.pyplot as plt
import os
import nltk
from wordcloud import WordCloud
```

```python
name=input("enter name with space\t ")
def make_wordcloud():
    wordcloud= WordCloud(width=480, height=480, margin=0).generate(name)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.margins(x=0,y=0)
    plt.show()
make_wordcloud()
```
